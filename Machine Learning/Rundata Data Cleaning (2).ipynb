```python
import pandas as pd
import numpy as np
```

```python

file_path = 'rundata.xlsx'
df = pd.read_excel(file_path, sheet_name='Sheet1')
```

```python
df.info()
```

```python
n = df.shape[0]
r = df['Bildlänk'].isnull().sum()
p = (r*100)//n
print(f'Null values in the "Bildlänk" column = [{r}] that counts for [{p}%].')
df = df.drop(columns=['Bildlänk'], axis=1)
print("""
      The column 'Bildlänk' has been dropped as it includes non-essential information, 
      high proportion of missing values, and dropping it improves the efficiency of data processing and model training.
      """)
```

```python
n = df.shape[0]
r = df['Referens'].isnull().sum()
p = (r*100)//n
print(f'Null values in the "Referens" column = [{r}] that counts for [{p}%].')
df = df.drop(columns=['Referens'], axis=1)
print("""
      The column 'Referens' has been dropped as it includes non-essential information, 
      contains unstructured text data, high proportion of missing values, and dropping it can simplify the dataset and potentially improve model performance.
      """)
```

```python
n = df.shape[0]
r = df['Alternativt signum'].isnull().sum()
p = (r*100)//n
print(f'Null values in the "Alternativt signum" column = [{r}] that counts for [{p}%].')
df = df.drop(columns=['Alternativt signum'], axis=1)
print("""
      The column 'Alternativt signum' has been dropped as it includes non-essential information, 
      alternativt signum are not going to be used in our specific analysis or prediction task.
      """)
```

```python
n = df.shape[0]
r = df['Övrigt'].isnull().sum()
p = (r*100)//n
print(f'Null values in the "Övrigt" column = [{r}] that counts for [{p}%].')
df = df.drop(columns=['Övrigt'], axis=1)
print("""
      The column 'Övrigt' has been dropped as it includes non-essential information, 
      missing data, free-text information (can be challenging when processed by machine learning).
      """)
```

```python

column_names = {
    'Signum': 'sigma',
    'Plats': 'plats',
    'Socken': 'socken',
    'Härad': 'harad',
    'Kommun': 'kommun',
    'Placering': 'placering',
    'Koordinater': 'koordinater',
    'Urspr. plats?': 'ursprunglig_plats',
    'Nuv. koord.': 'nuvarande_koordinater',
    'Sockenkod/Fornlämningsnr.': 'fornlamningsnummer',
    'Runtyper': 'runtyp',
    'Korsform': 'korsform',
    'Period/Datering': 'datering',
    'Stilgruppering': 'stilgrupp',
    'Ristare': 'ristare',
    'Materialtyp': 'materialtyp',
    'Material': 'material',
    'Föremål': 'foremal'
}

df.rename(columns=column_names, inplace=True)
print('New Dataset column names:\n')
for old, new in column_names.items():
    print(f'[{old}]: {new}')
```

```python
categorical_list = ['plats', 'socken', 'harad', 'kommun']

for col in categorical_list:
    df[col] = df[col].replace(r'^\?$', np.nan, regex=True)

df['harad'] = df['harad'].str.replace('?,', '')
```

```python
categorical_columns = [
    'sigma', 'plats', 'socken', 'harad', 'kommun', 'placering', 
    'ursprunglig_plats', 'runtyp', 'datering', 'stilgrupp', 'ristare', 
    'materialtyp', 'material', 'foremal'
]

print("Number of empty values (NaN) replaced with 'missing' in the dataset columns:")
for col in categorical_columns:
    nan_count = df[col].isnull().sum()
    print(f'[{col}]: [{nan_count}]') 
    df[col] = df[col].fillna('missing') 
```

```python

df['cleaned_foremal'] = df['foremal'].str.lower()
df['cleaned_foremal'] = df['cleaned_foremal'].str.replace(r'[0-9]', '', regex=True)
df['cleaned_foremal'] = df['cleaned_foremal'].str.replace('(?)', '')
df['cleaned_foremal'] = df['cleaned_foremal'].str.replace('[]', '')
df['cleaned_foremal'] = df['cleaned_foremal'].str.replace('()', '')
df['cleaned_foremal'] = df['cleaned_foremal'].str.replace('?', '')
df['cleaned_foremal'] = df['cleaned_foremal'].str.strip()


foremal_list = [
    'fragment av runsten', 'fragment av gravhäll', 'läderslida', 'laggskålsbotten', 
    'putsinskrift', 'dopfunt', 'runben, revben', 'runmålning', 'stavkyrkegrafitti', 'trästycke'
]

for item in foremal_list:
    df.loc[df['cleaned_foremal'].str.contains(item, case=False, na=False), 'cleaned_foremal'] = item

df.loc[df['cleaned_foremal'].str.contains('runben, revben, fragment', case=False, na=False), 'cleaned_foremal'] = 'runben, revben'
df.loc[df['cleaned_foremal'].str.contains('revben, fragment,  st', case=False, na=False), 'cleaned_foremal'] = 'revben, fragment'
```

```python
okänd_count_materialtyp = (df['materialtyp'] == 'okänd').sum()
print(f"Number of rows to be dropped with 'okänd' entry in 'materialtyp': [{okänd_count_materialtyp}]")

df = df.drop(df[df['materialtyp'] == 'okänd'].index)
```

```python
missing_count_materilatype = (df['materialtyp'] == 'missing').sum()
missing_count_datering = (df['datering'] == 'missing').sum()
print(f"Number of rows with 'missing' in 'materialtyp': [{missing_count_materilatype}] and in 'datering': [{missing_count_datering}]")

df.drop(df[(df['materialtyp'] == 'missing') & (df['datering'] == 'missing')].index, inplace=True)
```

```python

df['cleaned_material'] = df['material'].str.lower()
df['cleaned_material'] = df['cleaned_material'].str.replace('(?)', '')
df['cleaned_material'] = df['cleaned_material'].str.replace('?', '')
df['cleaned_material'] = df['cleaned_material'].str.strip()

material_list = ['brons', 'guld', 'silver', 'kalkputs', 'trä', 'amfibolit', 'ben', 'bok', 'ek', 'furu']

for item in material_list:
    df.loc[df['cleaned_material'].str.contains(item, case=False, na=False), 'cleaned_material'] = item

```

```python
n = df['korsform'].str.contains(r"\[Klassificerad av redaktionen\]").sum()
print(f'The "korsform" column include [{n}] entries with [Klassificerad av redaktionen] substring.')

df['korsform'] = df['korsform'].str.replace('[Klassificerad av redaktionen]','', regex=False)
print('Substring [Klassificerad av redaktionen] has been removed from the entries of the "korsform" column.')
print(f'After removal it includes [{df['korsform'].str.contains(r"\[Klassificerad av redaktionen\]").sum()}] entries.')
```

```python

df['cleaned_stilgrupp'] = df['stilgrupp']

stilgrupp_list = [
    '[Gräslund muntligt]', '[Klassificerad av redaktionen]', '[Källström 2007a:362]',
    '[Ljung muntligt]', '[NOR 2004:27]', '[Fv 2009]', '[Fv 2012]', '[Källström i SRI 10 Dig. suppl.]', ' [En av, Ljung muntligt]',
    ' [Källström 2007a:361]', ' [Källström 2007a:407]'
    ]

for item in stilgrupp_list:
    df['cleaned_stilgrupp'] = df['cleaned_stilgrupp'].str.replace(item, '')

df['cleaned_stilgrupp'] = df['cleaned_stilgrupp'].replace('sid a: Pr2, sid b: Pr2 - Pr3', 'missing')
df['cleaned_stilgrupp'] = df['cleaned_stilgrupp'].replace(r'^\?$', np.nan, regex=True)
df['cleaned_stilgrupp'] = df['cleaned_stilgrupp'].str.replace('?', '')

df['cleaned_stilgrupp'] = df['cleaned_stilgrupp'].str.strip()
df['cleaned_stilgrupp'] = df['cleaned_stilgrupp'].fillna('missing')
```
